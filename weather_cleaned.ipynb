```python
import pandas as pd
import numpy as np
```

```python
# Read the data from a CSV file
data = pd.read_csv('weather.csv')
data
```

```python
#missing values
missing_values = data.isnull().sum()
print("Missing values:")
print(missing_values)
```

```python
data_cleaned = data.dropna()
data_cleaned
```

```python
data_cleaned = data_cleaned.drop_duplicates()
string_columns = data_cleaned.select_dtypes(include=['object']).columns

data_cleaned
```

```python
string_columns
```

```python
data_cleaned[string_columns] = data_cleaned[string_columns].apply(lambda x: x.str.lower())
data_cleaned
```

```python
date_columns = ['Date']
data_cleaned[date_columns] = data_cleaned[date_columns].apply(pd.to_datetime)
data_cleaned
```

```python
data_cleaned.to_csv('weather_cleaned.csv', index=False)
```
