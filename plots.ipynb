```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Read the cleaned data from the CSV file
data = pd.read_csv('actual_weather.csv')

# Define the independent variable (predictor) and dependent variable (outcome)
X = data['Temp_C']  # Predictor variable (assuming 'Temp_C' is temperature)
y = data['Rel Hum_%']  # Outcome variable (assuming 'Rel Hum_%' is humidity)

# Fit the linear regression model
X = sm.add_constant(X)  # Add constant term to the predictor variable
model = sm.OLS(y, X).fit()

# Get the coefficients of the linear regression model
intercept, slope = model.params

# Plot the data points
plt.scatter(X['Temp_C'], y, color='blue', label='Data Points')

# Plot the regression line
plt.plot(X['Temp_C'], intercept + slope*X['Temp_C'], color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%)')
plt.title('Linear Regression: Temperature vs. Humidity')
plt.legend()

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Read the cleaned data from the CSV file
data = pd.read_csv('weather_cleaned.csv')

# Define the independent variables (predictors) and dependent variable (outcome)
X = data[['Temperature (C)', 'Rainfall (mm)', 'Wind Speed (m/s)']]  # Predictor variables
y = data['Humidity (%)']                                 # Outcome variable

# Fit the multiple regression model
X = sm.add_constant(X)  # Add constant term to the predictor variables
model = sm.OLS(y, X).fit()

# Predict the values using the model
y_pred = model.predict(X)

# Plot the relationship between observed and predicted values
plt.scatter(y, y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')  # Identity line
plt.xlabel('Observed Humidity')
plt.ylabel('Predicted Humidity')
plt.title('Multiple Regression: Observed vs Predicted Values')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Read the cleaned data from the CSV file
data = pd.read_csv('weather_cleaned.csv')

# Define the independent variable (predictor) and dependent variable (outcome)
X = data['Temperature (C)'].values.reshape(-1, 1)  # Predictor variable (reshaped for sklearn)
y = data['Humidity (%)'].values                    # Outcome variable

# Define the degree of the polynomial
degree = 2 # Change this to the desired degree of the polynomial

# Transforming the predictor variable to include polynomial terms up to degree
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Fit the polynomial regression model
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# Predict the values using the model
y_pred = poly_model.predict(X_poly)

# Sort the values of X before plotting
sorted_zip = sorted(zip(X, y_pred))
X_sorted, y_pred_sorted = zip(*sorted_zip)

# Plot the polynomial regression curve
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X_sorted, y_pred_sorted, color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
