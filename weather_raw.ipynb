```python
import pandas as pd
import numpy as np
from faker import Faker
import random
import string

```

```python
fake = Faker()
# Generate date ranges
date_range = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
# Generate sample data
# Generate sample data for weather
data = {
    'City': [fake.city() for _ in range(100)],
    'Date': [random.choice(date_range) for _ in range(100)],
    'Temperature (C)': [random.uniform(-20, 40) for _ in range(100)],  # Assuming temperature in Celsius
    'Humidity (%)': [random.randint(0, 100) for _ in range(100)],
    'Rainfall (mm)': [random.uniform(0, 50) for _ in range(100)],
    'Wind Speed (m/s)': [random.uniform(0, 20) for _ in range(100)]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
for col in random.sample(df.columns.tolist(), 3):
    df.loc[random.sample(range(len(df)), 5), col] = np.nan
# Add duplicates
df = pd.concat([df, df.sample(frac=0.1)])
df
```

```python
# Save to CSV
df.to_csv('weather.csv', index=False)
```
